```python
from rake_nltk import Rake
import pandas as pd
import numpy as np
```

```python
r = Rake()
```

```python
data = pd.read_csv('trainset.csv')
data['ref'] = data['ref'].apply(lambda x: x.lower())
data['mr'] = data['mr'].apply(lambda x: x.lower())
```

```python
n = 10
for i in range(n):
    MR = data.iloc[i]['mr']
    sentences = data.iloc[i]['ref']
    r.extract_keywords_from_text(sentences)
    print('---')
    print(MR)
    print(sentences)
    print(r.get_ranked_phrases())
```

```python
sen_lens = np.zeros(len(data['ref']))
for i in range(len(data['ref'])):
    sentence = data['ref'].iloc[i]
    sen_lens[i] = len(sentence.split())
```

```python
max(sen_lens)
```

```python
def clean_slot_name(slot):
    if slot == "pricerange":
        return "price range"
    if slot == "familyfriendly":
        return "family"
    if slot == "eattype":
        return "eat type"
```

```python
n = 300
for i in range(n):
    MR = data.iloc[i]['mr']
    sentences = data.iloc[i]['ref']
    MRs = MR.split(', ')
    pr = False
    for mr in MRs:
        slot = mr[:mr.find("[")]
        value = mr[mr.find("[")+1:mr.find("]")]
        #print(value)
        if value not in sentences or ((value == "yes" or value == "no") and clean_slot_name(slot) not in sentences):
            pr = True
    if pr:
        print(sentences)
        print(MRs)
        print('--')
```
